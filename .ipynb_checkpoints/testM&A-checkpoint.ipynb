```python
cities = ['chennai','bangalore','mumbai','pune','hyderabad','kolkata','coimbatore','delhi','gurgoan',
         'guwahati','bengaluru','ahmedabad','indore','surat','kanpur','jaipur','lucknow','nagpur',
         'patna','kochi','visakhapatnam','agra','allahabad']
for city in cities:
    urlbase = 'https://api.crunchbase.com/v/3/organizations?locations='+city+'&user_key=990526518aa6f95c9aff9e6b6ebc5c79'
    data =  requests.get(urlbase).json()
    total_pages = data['data']['paging']['number_of_pages']

    for page in range(1,int(total_pages)+1):
        url = urlbase+'&page='+str(page)
        print(url)
        print(len(orgs))
        data =  requests.get(url).json()
        for org in data['data']['items']:
            orgs.append(org)
```

```python
orgsdupRDD = sc.parallelize(orgs)
```
