```python
import scrapy
import json
from scrapy.crawler import CrawlerProcess
from scrapy.spiders import CrawlSpider
from scrapy.utils.project import get_project_settings
```

```python
from scrapy import optional_features
optional_features.remove('boto')
```

```python
class VcItem(scrapy.Item):
    # define the fields for your item here like:
    # name = scrapy.Field()
    title = scrapy.Field()
    desc = scrapy.Field()
    post = scrapy.Field()
    
class VcSpider(CrawlSpider):
	name = "vcposts"
	allowed_domains = ["www.vccircle.com"]
	start_urls = [
	"http://www.vccircle.com/deals-mna"
	]
	for page in range(32,43):
		start_urls.append('http://www.vccircle.com/deals-mna?page='+str(page))

   	def parse(self, response):
   		#print response.request.headers['User-Agent']
   		for href in response.xpath('//div[@class="BB"]/h2/a/@href'):
   			url = response.urljoin(href.extract())
   			yield scrapy.Request(url, callback=self.parse_dir_contents)
   			

   	def parse_dir_contents(self, response):
   		item = VcItem()
   		item['title'] = response.xpath('//div[@class="innercont-heading"]/h2/text()').extract()
   		item['desc'] = response.xpath('//div[@class="innercont-heading"]/div[@class="graycolor"]/text()').extract()
   		item['post'] = " ".join(response.xpath('//div[@class="cont-text"]//div//p/text()').extract())
   		with open('vcnote.json','a') as f:
   			json.dump({'title':item.get('title'),'desc':item.get('desc'),'post':item.get('post')},f)
        

```

```python
process = CrawlerProcess(get_project_settings())
process.crawl(VcSpider)
process.start()
```

```python
process.stop()
```
